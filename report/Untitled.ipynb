```python
import numpy as np
import pylab as pl

Y, X = np.mgrid[-10:10:100j, -10:10:100j]

def f(x, a, b):
    return x**3 + a*x + b

def root(x, a, b):
    return np.sqrt(np.abs(f(x, a, b)))

a = -2
b = 4

pl.contour(X, Y, Y**2 - f(X, a, b), levels=[0])
```

```python
x = np.arange(a, 10, 0.2)

y1 = root(x, a, b)
y2 = -root(x, a, b)
pl.plot(x, y1)
pl.plot(x, y2)
pl.xlim([-10, 10])
pl.ylim([-10, 10])
print(y1)
print(y2)
```

```python
n = 103
field = list(range(n))
right_side = [(i**3+7) % n for i in range(n)]
left_side = [i**2 % n for i in range(n)]

X = []
Y = []
points = []

for i in range(len(right_side)):
    I = [index for index in range(len(right_side)) if right_side[i] == left_side[index]]
    for j in range(len(I)):
        X.append(field[i])
        Y.append(field[I[j]])
        points.append([field[i], field[I[j]]])

print(X)
print(Y)
result = [f"({el[0]}, {el[1]})" for el in points]
print(' '.join(result))
```
